```python
#import libraries here
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
                   
df2=pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
```

### Create a dataframe (df3) which consists both dataframe inner-joined by column "Key" (5 pts)

```python
#penggunaan .merge digunakan untuk menggabungkan 2 DataFrame
#penggunaan how='inner' untuk inner join dari 2 DataFrame
df3 = pd.merge(df1,df2,how="inner")
df3
```

### Create a dataframe (df4) which consists of the result of df1 and df2 left-joined by column "Key" (10pts)

```python
#penggunaan .merge digunakan untuk menggabungkan 2 DataFrame
#penggunaan how='inner' untuk left join dari 2 DataFrame
df4 = pd.merge(df1,df2,how="left")
df4
```

#### 1. Replace key "K2" into "K3" on df1
#### 2. Left join it with df2 
(15pts)

```python
#untuk replace value key maka kita butuh panggil dataframe dengan keynya.
#fungsi .replace berfungsi untuk mengganti value pada sebuah data
# "inplace=True" untuk menyimpan perubahan pada DataFrame ketika melakukan rename value tanpa harus define dataframe baru
df1['key'].replace(['K2'], 'K3', inplace=True)
df1
```

```python
# ".merge" digunakan untuk menggabungkan 2 DataFrame
# "how" merupakan alokasi untuk melakukan join pada DataFrame
# "left" pada join yaitu sistem akan mengambil value dari tabel kiri
df1 = pd.merge(df1,df2,how="left")
df1
```

# Dataframe Processing


Create a dictionary and convert it to a dataframe called "Customer_df" (15 pts)


name=['Anna','Jason','Cindy']

age=['22','23','21']

gender=['F','M','F']




```python
#membuat dataframe baru dari dict dengan nama Customer_df
customer = { 
    'name':['Anna','Jason','Cindy'],
    'age':['22','23','21'],
    'gender':['F','M','F']
}
# pd.DataFrame adalah untuk convert dict tersebut menjadi dataframe
# untuk memanggil dict tersebut dapat diberi nama Customer_df
Customer_df = pd.DataFrame(customer)
Customer_df
```

### Delete the "age" column (5 pts)

```python
#menghapus kolom dengan .drop
#penggunaan axis =1 itu ke kolom kalau axis=0 atau default itu menghapus index
Customer_df = Customer_df.drop('age',axis=1)
Customer_df
```

### Rename the "name" column into "first name" (10 pts)

```python
#Cara 1 dengan inplace true, maka dataframe langsung dirubah dengan rename yang baru tanpa harus mendifine dataframe terlebih dulu
# ".rename" merupakan parameter untuk mengganti nama/value pada data
Customer_df.rename(columns={'name':'first_name'}, inplace = True)
Customer_df
```

```python
#Cara 2 tidak dengan inplace true, maka perlu didefine Customer_df terlebih dahulu
# ".rename" merupakan parameter untuk mengganti nama/value pada data
Customer_df = Customer_df.rename(columns={'name':'first_name'})
Customer_df
```

### Sort the dataframe by age, descending (10 pts)

```python
#mengurutkan atau sorting by age secara descending dengan .sortvalues dan 'age'. lalu untuk descending kita dapat menggunakan ascending = false maka akan dibaca sebagai kebalikannya atau descending
#.reset_index digunakan untuk mereset index dari 0 
# drop=True merupakan sebuah parameter yang digunakan untuk menghindari indeks lama ditambahkan sebagai kolom
customer = { 
    'name':['Anna','Jason','Cindy'],
    'age':['22','23','21'],
    'gender':['F','M','F']
}
Customer_df = pd.DataFrame(customer)
Customer_df.sort_values(by=['age'], ascending=False).reset_index(drop=True)
```

## Using data aggregation in pandas , Answer the question below:

Show the total amount of passengers who survived and not survived (10pts)

```python
#passenger_survived_df = df.groupby(['Survived'])['PassengerId'].count()
#".map" berfungsi untuk mengganti setiap nilai dalam rangkaian dengan nilai lain
#".groupby" berfungsi untuk mengelompokan sesuai dengan data yang ingin dikelompokan
#".agg" berfungsi sebagai proses dimana data mentah dikumpulkan dan diekspresikan sabagai summary dan .agg hanya bisa dilakukan dengan .groupby 
#"count" guna untuk menjumlahkan row pada kolom 
df['Survived']=df.Survived.map({0:'Not Survived',1:'Survived'})
passenger_survived_df = df.groupby(['Survived']).agg({'PassengerId':['count']})
passenger_survived_df
```

Show all male survivors from titanic incident (10 pts)

```python
#untuk menampilkan male survivor menggunakan filtering data
male_survivor_df = df[(df['Sex']=='male')&(df['Survived']==1)]
male_survivor_df
```

Show the average fare paid based on gender (10 pts)

```python
#menggunakan .groupby dan .agg untuk menampilkan rata-rata fare paid berdasarkan gender
average_fare_df = df.groupby(['Sex']).agg({'Fare':['mean']})
average_fare_df
```
